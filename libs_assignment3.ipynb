# Question 1- Rice Market Analysis (NumPy)

```python
import numpy as np

# a) Create NumPy array
prices = np.array([1200, 1250, 1180, 1300, 1220, 1270, 1240])
print("Array:", prices)

# b) Mean price
mean_price = np.mean(prices)
print("Mean price:", mean_price)  # 1237.14

# c) Highest and lowest prices
highest = np.max(prices)
lowest = np.min(prices)
print("Highest:", highest, "Lowest:", lowest)  # 1300, 1180

# d) Difference between highest and lowest
price_range = highest - lowest
print("Price range:", price_range)  # 120
```

# Question 2 - Price Shock Investigation (NumPy)

```python
# Original prices
prices = np.array([1200, 1250, 1180, 1300, 1220, 1270, 1240])
old_mean = np.mean(prices)

# a) New prices with 2000 instead of 1240
new_prices = np.array([1200, 1250, 1180, 1300, 1220, 1270, 2000])
new_mean = np.mean(new_prices)

print("Old mean:", old_mean)  # 1237.14
print("New mean:", new_mean)  # 1345.71

# b) Compare
print("Mean increased by:", new_mean - old_mean)  # 108.57
```

# Question 5 - Visualizing School Attendance (Matplotlib)

```python
import matplotlib.pyplot as plt

attendance = [35, 37, 40, 38, 42, 39, 41]
days = range(1, 8)

plt.figure(figsize=(8,5))
# a) Line graph + c) markers
plt.plot(days, attendance, marker='o', linestyle='-', color='blue')

# b) Title and axis labels
plt.title('School Attendance Over 7 Days')
plt.xlabel('Day')
plt.ylabel('Number of Students')
plt.grid(True)
plt.xticks(days)
plt.show()
```
