# Problem 10

```python
original_price = 2500  # Original price
discount_rate = 0.10  # 10% as a decimal

discount_amount = original_price * discount_rate  # Calculate discount
final_price = original_price - discount_amount  # Subtract  discount from original price

print("Discount amount:", discount_amount, "LRD")
print("Final price after discount:", final_price, "LRD")
```
