# Question 1: Type Conversion (The Nimba Truck Report)

```python

transport_cost = "750"  # This is text because of the quotes
transport_cost_int = int(transport_cost)  # Convert text "750" to integer 750
car_loader_fee = 250  # Fee in LRD
total_delivery_cost = transport_cost_int + car_loader_fee  # Add them together

print(f"Ma Christiana, your final shipping cost is:{total_delivery_cost} LRD.")
```

#Explanation:  
The driver sent "750" as text, so Python sees it as a string, not a number. We use int() to convert it into a real number we can do math with. After adding the 250 LRD fee, we store the result in total_delivery_cost. The f-string lets us insert that number directly into the sentence without breaking the text.

# Question 4: Methods vs. Attributes (The Commercial Transport Truck)

```python
import pandas as pd

gobachop_data = pd.DataFrame({
    "Vendor": ["Ma Musu", "Sia", "Flomo"],
    "Sales": [4000, 9500, 1500]
})

# Example of attributes: .shape tells us the size of the DataFrame
print("DataFrame shape:", gobachop_data.shape)

# Example of method: .head() shows the first few rows
print(gobachop_data.head())

# Example of method: .describe() gives statistics for numeric columns
print(gobachop_data.describe())
```

# Explanation:  
In pandas, an attribute is a property that stores data about the DataFrame, like .shape. You don’t call it with (). A method is an action the DataFrame can perform, like .head() or .describe(), so you call it with parentheses. Attributes describe, methods do.
